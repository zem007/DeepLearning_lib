### train DeepGCN model

```python
import sys
sys.path.append('/home/maze/USAL/libs/')
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import math
import random
import time
from models.tensorflow_models.deep_gcns import DeepGCNs
from sklearn import preprocessing
from evaluation.metrics_tf import dice_coefficient
from utils.tf_utils import shuffle_mini_batches

BASE_LEARNING_RATE = 0.001
BATCH_SIZE = 8
DECAY_STEP = 300000
DECAY_RATE = 0.5
BN_INIT_DECAY = 0.5
BN_DECAY_DECAY_RATE = 0.5
BN_DECAY_DECAY_STEP = float(DECAY_STEP)
BN_DECAY_CLIP = 0.99
    
def get_learning_rate(batch):
    learning_rate = tf.train.exponential_decay(
                        BASE_LEARNING_RATE,  # Base learning rate.
                        batch * BATCH_SIZE,  # Current index into the dataset.
                        DECAY_STEP,          # Decay step.
                        DECAY_RATE,          # Decay rate.
                        staircase=True)
    learning_rate = tf.maximum(learning_rate, 0.00001) # CLIP THE LEARNING RATE!
    return learning_rate        

def get_bn_decay(batch):
    bn_momentum = tf.train.exponential_decay(
                      BN_INIT_DECAY,
                      batch*BATCH_SIZE,
                      BN_DECAY_DECAY_STEP,
                      BN_DECAY_DECAY_RATE,
                      staircase=True)
    bn_decay = tf.minimum(BN_DECAY_CLIP, 1 - bn_momentum)
    return bn_decay

def convert_to_cls_labels(y_array):
    labels_list = []
    for i in range(len(y_array)):
        if max(y_array[i]) == 1 or max(y_array[i]) == 0:
            labels_list.append(0)
        elif max(y_array[i]) == 2:
            labels_list.append(1)
        else:
            print(max(y_array[i]))
            print(i)
    labels_array = np.array(labels_list)
    return labels_array

def convert_to_seg_labels(y_array):
    assert(len(y_array.shape) == 2)
    for i in range(y_array.shape[0]):
        for j in range(y_array.shape[1]):
            if y_array[i,j] == 2 or y_array[i,j] == 1:
                y_array[i,j] = 1

    return y_array
```

```python
# predefine something
which_gpu = '1'
data_saver_filename = './df_1024_triangle.pkl'
saver_folder_name = '/home/maze/USAL/libs/model_saver'
num_class = 2
batch_size = 8
epochs = 150
learning_rate = 0.001
momentum = 0.9
shuffle = True
with_aug = False
with_adj = False
OPTIMIZER = 'adam'
GCN = 'edgeconv'

# Training Settings
NUM_POINTS = 1024
NUM_LAYERS = 28

# GCN parameters
NUM_NEIGHBORS = [16]
if (len(NUM_NEIGHBORS) < NUM_LAYERS):
    while (len(NUM_NEIGHBORS) < NUM_LAYERS):
        NUM_NEIGHBORS.append(NUM_NEIGHBORS[-1]) 

NUM_FILTERS = [64]
if (len(NUM_FILTERS) < NUM_LAYERS):
    while (len(NUM_FILTERS) < NUM_LAYERS):
        NUM_FILTERS.append(NUM_FILTERS[-1]) 

DILATIONS = [-1]
if DILATIONS[0] < 0:
    DILATIONS = [1] + list(range(1, NUM_LAYERS))
elif (len(DILATIONS) < NUM_LAYERS):
    while (len(DILATIONS) < NUM_LAYERS):
        DILATIONS.extend(DILATIONS) 
    while (len(DILATIONS) > NUM_LAYERS):
        DILATIONS.pop() 

tf.reset_default_graph()
os.environ["CUDA_VISIBLE_DEVICES"] = which_gpu
# read data
df = pd.read_pickle(data_saver_filename)
df = df.drop(['Rupture'], axis = 1)
print('the original df len is: ', len(df))
# to array
df_train = df[df.data_set == 'train_set']
df_val = df[df.data_set == 'validation_set']

x_train = df_train.iloc[:, -3].tolist()
x_train = np.array(x_train)
print('x_train shape is ', x_train.shape)
labels_train = df_train.iloc[:, -2].tolist()
labels_train = np.array(labels_train)
print('labels_train shape is ', labels_train.shape)
adj_train = df_train.iloc[:, -1].tolist()
adj_train = np.array(adj_train)
print('adj_train shape is ', adj_train.shape)
info_train = df_train.iloc[:, 4: -3].values
minmaxscaler = preprocessing.MinMaxScaler()
info_train = minmaxscaler.fit_transform(info_train)
print('info_train shape is ', info_train.shape)

x_val = df_val.iloc[:, -3].tolist()
x_val = np.array(x_val)
print('x_validation shape is ', x_val.shape)
labels_val = df_val.iloc[:, -2].tolist()
labels_val = np.array(labels_val)
print('labels_val shape is ', labels_val.shape)
adj_val = df_val.iloc[:, -1].tolist()
adj_val = np.array(adj_val)
print('adj_validation shape is ', adj_val.shape)
info_val = df_val.iloc[:, 4: -3].values
info_val = minmaxscaler.transform(info_val)
print('info_validation shape is ', info_val.shape)
```

```python
temp_best_acc = 0
labels_train_cls = convert_to_cls_labels(labels_train)
labels_train_seg = convert_to_seg_labels(labels_train)
labels_val_cls = convert_to_cls_labels(labels_val)
labels_val_seg = convert_to_seg_labels(labels_val)
        
num_train_sample = x_train.shape[0]
num_val_sample = x_val.shape[0]
num_point = x_train.shape[1]  # num_point == num_vertexes
num_coordinate = 3 # for this DeepGCNs
nums_info = info_train.shape[1]
        
graph = tf.Graph()
with graph.as_default():
    
    x = tf.placeholder(tf.float32, shape=(batch_size, num_point, num_coordinate))
    labels_seg = tf.placeholder(tf.int32, shape=(batch_size, num_point))
    labels_cls = tf.placeholder(tf.int32, shape=(batch_size))
    training_flag = tf.placeholder(tf.bool, shape=())
    info = tf.placeholder(tf.float32, shape=(batch_size, nums_info))
    if with_adj == True:
        adjs = tf.placeholder(tf.float32, shape=(batch_size, num_point, num_point))
    else: adjs = False
    
    batch = tf.Variable(0, trainable = False)
    bn_decay = get_bn_decay(batch)
    learning_rate = get_learning_rate(batch)
    # define optimizer
    if OPTIMIZER == 'momentum':
        optimizer = tf.train.MomentumOptimizer(learning_rate, momentum=momentum)
    elif OPTIMIZER == 'adam':
        optimizer = tf.train.AdamOptimizer(learning_rate)
        
    # Configure the neural network using every layers
    deep_gcns_init = DeepGCNs(inputs = x,
                              is_training = training_flag,
                              kernel_size = [1,1], 
                              padding = 'VALID', 
                              stride = [1,1], 
                              bn=True,
                              weight_decay=0.0,
                              bn_decay=bn_decay, 
                              is_dist = True, 
                              num_point = NUM_POINTS,
                              num_layers = NUM_LAYERS,
                              num_neighbors = NUM_NEIGHBORS,
                              num_filters = NUM_FILTERS,
                              num_class = num_class,
                              info_arr = info,
                              adjs = adjs, 
                              skip_connect='residual',
                              dilations=DILATIONS)

    # build model and define losses
    pred_seg = deep_gcns_init.pred
    pred_cls = deep_gcns_init.pred_cls
    loss_seg = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(logits=pred_seg, labels=labels_seg))
    loss_cls = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(logits=pred_cls, labels=labels_cls))
    total_loss = loss_seg + loss_cls
    dice = dice_coefficient(pred_seg, labels_seg)
            
    train_op = optimizer.minimize(total_loss, global_step=batch)
    # Add ops to save and restore all the variables.
    saver = tf.train.Saver(max_to_keep = 1)
    config = tf.ConfigProto()
    config.gpu_options.allow_growth = True
    config.allow_soft_placement = True
    config.log_device_placement = True
# open a new session
with tf.Session(graph = graph, config = config) as sess:
    init = tf.global_variables_initializer()
    sess.run(init)
    print('----------------start training-----------------')
    t0 = time.time()
    for epoch in range(epochs):
        total_correct_cls = 0
        total_correct_seg = 0
        total_dice = 0
        total_seen_graph = 0
        total_seen_points = 0
        loss_sum_seg = 0
        loss_sum_cls = 0
        loss_sum = 0

        num_minibatches = math.ceil(num_train_sample/batch_size)
        # random shuffle the whole train data set
        shuffled_minibatch_list = shuffle_mini_batches(x_train, labels_train_cls, labels_train_seg, info_train, batch_size, adj_train)
        for i in range(len(shuffled_minibatch_list)):
            one_batch = shuffled_minibatch_list[i]
            minibatch_x, minibatch_y_cls, minibatch_y_seg, minibatch_info, minibatch_adj = one_batch
            if with_aug == True:
                #data aug, rotate...
                rotated_point = rotate_point_cloud(minibatch_x)
                jittered_point = jitter_point_cloud(rotated_point)
                minibatch_x = shift_point_cloud(jittered_point)
#             feed_dict = {x: minibatch_x, labels_cls: minibatch_y_cls, labels_seg: minibatch_y_seg, 
#                         info: minibatch_info, adjs: minibatch_adj, training_flag: True}
            feed_dict = {x: minibatch_x, labels_cls: minibatch_y_cls, labels_seg: minibatch_y_seg, 
                        info: minibatch_info, training_flag: True}

#             _, minibatch_loss, minibatch_pred_seg, minibatch_dice = sess.run([train_op, total_loss, pred_seg, dice], feed_dict = feed_dict)
            _, minibatch_loss, minibatch_loss_cls, minibatch_loss_seg, minibatch_pred_cls, minibatch_pred_seg, minibatch_dice = sess.run(
                [train_op, total_loss, loss_cls, loss_seg, pred_cls, pred_seg, dice], feed_dict = feed_dict)

            pred_cls_output = np.argmax(minibatch_pred_cls, 1)
            correct_cls = np.sum(pred_cls_output == minibatch_y_cls)
            total_correct_cls += correct_cls
            total_seen_graph += batch_size
            pred_seg_output = np.argmax(minibatch_pred_seg, 2)
            correct_seg = np.sum(pred_seg_output == minibatch_y_seg)
            total_correct_seg += correct_seg
            total_seen_points += (batch_size * num_point)
            total_dice += minibatch_dice
            loss_sum_cls += minibatch_loss_cls
            loss_sum_seg += minibatch_loss_seg
            loss_sum += minibatch_loss

        if epoch % 1 == 0:
            print ("total loss after epoch %i: %f" % (epoch, (loss_sum/float(num_minibatches))))
            print ("seg loss after epoch %i: %f" % (epoch, (loss_sum_seg/float(num_minibatches))))
            print ("cls loss after epoch %i: %f" % (epoch, (loss_sum_cls/float(num_minibatches))))
            print ("train_accuracy for seg after epoch %i: %f" % (epoch, (total_correct_seg/float(total_seen_points))))
            print ("train_mean_dice after epoch %i: %f" % (epoch, (total_dice/float(num_minibatches))))
            print ("train_accuracy for cls after epoch %i: %f" % (epoch, (total_correct_cls/float(total_seen_graph))))
            print ("-----train end------")
            
        # start evaluate validation set
        total_correct_cls = 0
        total_correct_seg = 0
        total_dice = 0
        total_seen_graph = 0
        total_seen_points = 0
        loss_sum_seg = 0
        loss_sum_cls = 0
        loss_sum = 0

        num_minibatches_test = math.ceil(num_val_sample/batch_size) # number of minibatches in the validation set
        # random shuffle the whole train data set
        shuffled_minibatch_list = shuffle_mini_batches(x_val, labels_val_cls, labels_val_seg, info_val, batch_size, adj_val)
        for i in range(len(shuffled_minibatch_list)):
            one_batch = shuffled_minibatch_list[i]
            minibatch_x, minibatch_y_cls, minibatch_y_seg, minibatch_info, minibatch_adj = one_batch
#             feed_dict = {x: minibatch_x, labels_cls: minibatch_y_cls, labels_seg: minibatch_y_seg,
#                         info: minibatch_info, adjs: minibatch_adj, training_flag: False}
            feed_dict = {x: minibatch_x, labels_cls: minibatch_y_cls, labels_seg: minibatch_y_seg,
                        info: minibatch_info, training_flag: False}

            minibatch_loss, minibatch_loss_cls, minibatch_loss_seg, minibatch_pred_cls, minibatch_pred_seg, minibatch_dice = sess.run(
                        [total_loss, loss_cls, loss_seg, pred_cls, pred_seg, dice], feed_dict = feed_dict)

            pred_cls_output = np.argmax(minibatch_pred_cls, 1)
            correct_cls = np.sum(pred_cls_output == minibatch_y_cls)
            total_correct_cls += correct_cls
            total_seen_graph += batch_size
            pred_seg_output = np.argmax(minibatch_pred_seg, 2)
            correct_seg = np.sum(pred_seg_output == minibatch_y_seg)
            total_correct_seg += correct_seg
            total_seen_points += (batch_size * num_point)
            total_dice += minibatch_dice
            loss_sum_cls += minibatch_loss_cls
            loss_sum_seg += minibatch_loss_seg
            loss_sum += minibatch_loss

        epoch_test_accuracy = total_correct_cls/float(total_seen_graph)
        if epoch % 1 == 0:
            print ("validation total loss after epoch %i: %f" % (epoch, (loss_sum/float(num_minibatches_test))))
            print ("validation seg loss after epoch %i: %f" % (epoch, (loss_sum_seg/float(num_minibatches_test))))
            print ("validation cls loss after epoch %i: %f" % (epoch, (loss_sum_cls/float(num_minibatches_test))))
            print ("val_accuracy for seg after epoch %i: %f" % (epoch, (total_correct_seg/float(total_seen_points))))
            print ("validation_mean_dice after epoch %i: %f" % (epoch, (total_dice/float(num_minibatches_test))))
            print ("val_accuracy for cls after epoch %i: %f" % (epoch, (total_correct_cls/float(total_seen_graph))))
            print('---------------------------------------------------')

            # save the best model only
            if epoch_test_accuracy >= temp_best_acc:
                saver.save(sess = sess, save_path = saver_folder_name + '/Deep_Res_GCNS_weights.ckpt', global_step = epoch)
                temp_best_acc = epoch_test_accuracy

    print('----------------total training time is: ', str(time.time()- t0))
    print("Finished!")
```

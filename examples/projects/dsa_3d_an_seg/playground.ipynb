```python
import sys
sys.path.append('/home/maze/USAL/libs/')
import os
import numpy as np
from data_io.numpy_io import NumpyIo
from mapper.keras_generator import keras_image_generator
from keras.utils import np_utils

data_save_path = '/home/maze/USAL/libs/new_data'
# load dataset
numpyIo = NumpyIo()
x_train = numpyIo.load(path = os.path.join(data_save_path, 'x_train.npy'))
labels_train = numpyIo.load(path = os.path.join(data_save_path, 'labels_train.npy'))
x_test = numpyIo.load(path = os.path.join(data_save_path, 'x_val.npy'))
labels_test = numpyIo.load(path = os.path.join(data_save_path, 'labels_val.npy'))

def image_generator_2Dto3D(data, label, datagen, batch_size = 8, shuffle=True):  #Input dim (n,64,64,64,1)
    datagen.fit(data)
    for data_batch, label_batch in datagen.flow(data, label, batch_size=batch_size, shuffle = shuffle):
        yield(np.expand_dims(data_batch, axis=-1), np_utils.to_categorical(label_batch, 3))
```

```python
keras_datagen = keras_image_generator()
gen = image_generator_2Dto3D(data = x_train, label = labels_train, datagen = keras_datagen)
```

### 训练模型(第二版)

```python
from model.keras_model.unet3D import Unet3D

# class initialize
unet3d_class = Unet3D(which_gpu = '0', 
                      n_labels = 3,
                      input_shape = (64,64,64,1),
                      pool_size = (2,2,2), 
                      deconvolution = True,
                      depth = 4,
                      n_base_filters = 32,
                      batch_normalization = True,
                      activation_name = 'softmax',
                      model_save_path = '/home/maze/USAL/libs/model_save_path')

# build model structure
built_model = unet3d_class.build_model()
# compile model
compiled_model = unet3d_class.compile_model(built_model = built_model, 
                                            include_label_wise_dice_coefficients = True, 
                                            initial_learning_rate = 1e-5, 
                                            metrics = 'weighted_dice_coefficient', 
                                            loss_fun = 'weighted_dice_coefficient_loss')
# callbacks
model_callbacks = unet3d_class.callbacks(monitor = 'val_loss', save_best_only = True)
# train model
unet3d_class.train_model(compiled_model = compiled_model, 
                         model_callbacks = model_callbacks, 
                         x_train = x_train, 
                         labels_train = labels_train, 
                         x_test = x_test, 
                         labels_test = labels_test, 
                         using_data_augmentation = False, 
                         batch_size = 8, 
                         epochs = 100, 
                         verbose = 1, 
                         shuffle = True)
```

```python
from model.keras_model.unet3D import Unet3D
import numpy as np

# class initialize
unet3d_class = Unet3D(which_gpu = '0', 
                      n_labels = 3,
                      input_shape = (64,64,64,1),
                      pool_size = (2,2,2), 
                      deconvolution = True,
                      depth = 4,
                      n_base_filters = 32,
                      batch_normalization = True,
                      activation_name = 'softmax',
                      model_save_path = '/home/maze/USAL/libs/model_save_path')

# build model structure
built_model = unet3d_class.build_model()
# compile model
compiled_model = unet3d_class.compile_model(built_model = built_model, 
                                            include_label_wise_dice_coefficients = True, 
                                            initial_learning_rate = 1e-5, 
                                            metrics = 'weighted_dice_coefficient', 
                                            loss_fun = 'weighted_dice_coefficient_loss')
# get test img
test_img = x_test[0]
test_img = np.expand_dims(test_img, axis = -1)
test_img = np.expand_dims(test_img, axis = 0) # shape = (1,64,64,64,1)
# predict
pred_mask = unet3d_class.predict(compiled_model = compiled_model, test_img = test_img)
print(pred_mask.shape)
```
